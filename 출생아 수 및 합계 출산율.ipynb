```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15   #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = df.T #df.T 를 df에 입력
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(15, 10))
ax1.plot(df.index, df['출생아 수'], color = '#FF4500')

ax2 = ax1.twinx() #축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color = '#FFFF00')
```

```python
fig, ax1 = plt.subplots(figsize=(15, 7))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250 , 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color = '#FF8C00')

ax2 = ax1.twinx() #축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0 , 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color = '#FFFF00')
```

```python
fig, ax1 = plt.subplots(figsize=(15, 7))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250 , 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color = '#FF8C00')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 20, val, ha ='center')
    
ax2 = ax1.twinx() #축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color = '#FFFF00')
```

```python
fig, ax1 = plt.subplots(figsize=(15, 7))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250 , 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color = '#FF8C00')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 20, val, ha ='center')
    
ax2 = ax1.twinx() #축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color = '#FFFF00', marker = 'o' ,ms = 15, lw= 6, mec = 'w', mew = 5)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha = 'center')
plt.title('출생아 수 및 합계 출산율')
plt.savefig('출생아 수 및 합계 출산율.png')
plt.show()
```
